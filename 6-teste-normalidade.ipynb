```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt

# Verificação 1 - histograma

dados = norm.rvs(size=10000)
plt.hist(dados, bins=20)
plt.title("Dados")
plt.show()

# através do histograma, posso verificar visualmente se os dados estão
# normalmente distribuídos se o formato do histograma se aproxima do formato
# de uma distribuição normal.
```

```python
# Verificação 2 - gráfico de distribuição normal

fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()

# Se os pontos se alinham à linha diagonal, então os dados
# estão normalmente distribuídos. É normal haver divergências nas bordas.
```

```python
# Verificação 3 - teste de shapiro-wilk

stats.shapiro(dados)
# Se P > 0,05, então os dados estão normalmente distribuídos
```

```python
# Refazer as 3 verificações, mas dessa vez com dados não normais

dados2 = skewnorm.rvs(4, size=1000)
plt.hist(dados2, bins=20)
plt.title('Dados')
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)

# Se P > 0,05, então os dados estão normalmente distribuídos
```
