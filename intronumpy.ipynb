```python
import numpy as np
import matplotlib.pyplot as plt

# preciso excluir a primeira coluna desse arquivo, pois contém strings.
# esse dataset contém colunas de JAN2013 até MAR2020. 12 * 7 + 3 = 87
# preciso pegar da 2ª até a 87ª coluna, ou seja, range [1,2,...,87,88]

# no primeiro parâmetro também posso passar uma URL.
apples = np.loadtxt(
    "numpy-dados/apples_ts.csv", delimiter=",", usecols=np.arange(1, 88)
)

# apples.ndim -> 2 -> temos duas dimensões, só linhas e colunas
# apples.shape -> (6,87) -> 6 linhas e 87 colunas.
# apples.size -> 522 -> a tabela tem 522 elementos
# apples.dtype -> float64
# apples.itemsize -> 8 -> cada elemento ocupa 8 bytes
# apples.nbytes -> 4176 -> 522 * 8 = 4176

# Podemos realizar a transposição dessa tabela para obter 87 linhas e 6 colunas.
# A transposição vai facilitar a visualização da tabela.
apples = apples.T

# já que a tabela foi transposta, agora as datas estão na peimeira coluna.
# porém as datas estão em um formato estranho (mês.ano), seria melhor um inteiro.
datas = np.arange(1,88)

# os preços estão em todas as colunas estantes, da 2ª até a 5ª.
precos = apples[:,1:6]

moscow = precos[:,0]
#moscow_ano = np.array(moscow[0,12])
kaliningrad = precos[:,1]
petersburg = precos[:,2]
krasnodar = precos[:,3]
ekaterinburg = precos[:,4]
```
